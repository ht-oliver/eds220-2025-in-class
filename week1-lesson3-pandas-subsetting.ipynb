## Selecting a single column

Simplest case: selecting a single column by column name.

General syntax: 
```
df [`column_name`]
```

This is an example of **label-based subsetting**. Which means we select data from our data frame using the *names* of the columns, *not their position*.

### Example

````python
# Select number of bird species observed at Mugu Lagoon in spring
mul_spring = df['MUL_spring']

mul_spring
````

````python
type(mul_spring)
````

We can do label-ased subsetting of a single column using attritube syntax

`df.column_name`

### Example

````python
df.MUL_spring
````

Favor `df['column_name']` instead of `df.column_name`

Why?
Because your column name could be the same as an attribute, or a space in it.

## Selecting multiple columns...

... using a list of column names.

Syntax :
```
df[ [`column_1`, `column_2`, `column_13`] ]
```

Notice: there are double square brackets! This is because we are passing a list of names `[`column_1`, `column_2`, `column_13`]` through the selection brackets `[]`

````python
# Select species abundance in Tijuana Estuary during winter and fall
tje_wf = df[['TJE_winter', 'TJE_fall']]
````

````python
# Check the data type of tje_wf - I think it's going to be a data frame
print(type(tje_wf))

# Check the shape of the selection - 
print(tje_wf.shape)
````

## .. using a slice

To select a slice of the columns we will use a special case of `loc` selection.

Syntax:
```
df.loc[ : , 'column_start':'column_end']
```

`column_start` and `column_end` = starting and ending point of column slice, clice includes both endpoints.

- the first value passed to `loc` is used to select rows, using a `:` as the row-selection parament means "select all the rows"

### Example 

````python
# Select columns between CSM_winter and MUL_fall
csm_mul = df.loc[ : , 'CSM_winter':'MUL_fall']

csm_mul.head()
````

## Selecting rows...

### ... Using a condition

Syntax:
```
df[condition_on_rows]
```

that `condition_on_rows` can be many things.

### Example

We are interested in all data after 2020

````python
# Select all rows with year > 2020
post_2020 = df[df['year']>2020]
post_2020
````

condition for our rows = `df['year']>2020` this is a `pandas.Series` with **boolean values** (True of False) indicates which rows satisfy the condition year>2020

````python
# Check the type of df[`year`]>2020
print(type(df['year']>2020))

# Print the boolean series
df['year']>2020
````

### Example

Suppose we want to look at data from 2012 to 2015 (including both years). 

````python
subset = df[ df['year'].between(2012, 2015)]
subset
````

- `df['year']` = column with year values, this is a `pandas.Series`
- `df['year'].between()` we have that `between()` is a method for the pandas.Series. Calling it using `.`
- (2012, 2015): parameters. If we look at the documentation, we can see that the method includes both endpoint.
- `df['year'].between(2012,2015)` = a pandas.Series of boolean values indicating which rows have year equal to 2012, 2013, 2014, 2015
- when you put `df['year'].between(2012,2015)` in the selection brackets, we obtain the rows we need.

## Avoid using `loc` for selecting only rows.

It is equivalent to write

```
# Select rows with year<2015
df[df['year']<2015]
```

and

```
df.loc[df['year']<2015, :]
```

In the second one:
- using `df.loc[ row-selection, column-selection]`
- the row selection is the condition `df['year']<2015`
- the column selection is `:` = all columns

## ... using multiple conditions

We can combine conditions to select rows by surrounding each in parenthiese `()` and using the or operator `|` and the and operator `&`.

### Example: or

````python
# Select rows in CSM with more than 50 species in winter or fall
df[(df['CSM_winter']>50) | (df['CSM_fall']>50)]
````

Select rows in which both CSM and SDW have more that 60 bird species during spring

````python
df[(df['CSM_spring']>60) & (df['SDW_spring']>60)]
# Nothing to see there
````

### Selecting rows by position

Sometimes we may want to select rows based on their *actual position* in the data frame.
This is position-based subsetting. To do this, we use `iloc`

`df.iloc[row-indices]`

````python
# Select the fifth row (index=4)
df.iloc[4]
````

## Select rows and columns simultaneously 

Can be done using `loc` (labels) or `iloc` (positions)

### ... by labels or conditions

Syntax:
```
df.loc[row-selection, column-selection]
```

### Example
Select winter surveys for Mugu Lagoon and Tijuana Estuary after 2020

````python
df.loc[ df['year']>2020, ['MUL_winter', 'TJE_winter']]
````

### ... by position
Use `iloc`

Syntax:
```
df.ioc[row-indices, column-indices]
```

### Example
Suppose we want to select rows 3-7 (including 7) and fourth and fifth columns


````python
df.iloc[3:8, [3,4]]
````

## `iloc` column slection?

### Example

Suppose we want to access the 9th columns in `df`.
In this case, we already know this column is the San Dieguito Wetland during spring.

````python
df.iloc[:,8]
````

### Why wouldn't you want to do this?
This code is less reproducible than locating by names. If the data shape changes, you could get different results.

Also, it's more specific/intuitive when you locate by names.
